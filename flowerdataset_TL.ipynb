```python
import tflearn.datasets.oxflower17 as oxflower17
X, Y = oxflower17.load_data(one_hot=True)
```

Preprocess the data 

```python
X.shape
```

```python
IMAGE_SIZE = [224, 224] #expecting in 224,224
```

```python
vgg = VGG19(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
#don't include top layer
```

```python
for layer in vgg.layers:
    layer.trainable = False #Freezing the network
```

```python
x = Flatten()(vgg.output) #flattening for going through fully connected layers
```

```python
prediction = Dense(17, activation='softmax')(x)
```

```python
model = Model(inputs=vgg.input, outputs=prediction)
```

Summary of the model

```python
model.summary()
```

```python
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

Model training 

```python
model.fit(X, Y, batch_size=128, epochs=2, verbose=1)
```
